```python
def categorize_rating(rating):
    "Categorize ratings into high or low."""
    if rating >= 4.0:
        return "High"
    else:
        return "Low"
```

```python
vectorised_func= np.vectorize(categorize_rating)
```

```python
vectorised_func([1,2,3,4,5,6,7])
```
